# **Computing BLEU SCORE**

```python
from nltk.translate.bleu_score import corpus_bleu
```

```python
# Compute BLEU score
def compute_BLEU_score(output, ground_truth):
    bleu_score = corpus_bleu([[translation] for translation in ground_truth], output)
    print("BLEU Score:", bleu_score)
```

```python
def translate_from_source_to_target(source_lang,target_lang,source_sentences):
    # Translate Gujarati sentences to English
    output_sentences = []
    tokenizer = AutoTokenizer.from_pretrained(
        "facebook/nllb-200-distilled-600M", src_lang=source_lang
    )
    i=0
    for sentence in source_sentences:
        # print(sentence)
        print(i)
        i=i+1
        inputs = tokenizer(sentence, return_tensors="pt")

        translated_tokens = model.generate(
            **inputs, forced_bos_token_id=tokenizer.lang_code_to_id[target_lang], max_length=30
        )
        translated_sentence=tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
    #     print(translated_sentence)
        output_sentences.append(translated_sentence)
    return output_sentences
```

```python
import torch
from transformers import AutoModelForSeq2SeqLM, BitsAndBytesConfig
from IndicTransTokenizer import IndicProcessor, IndicTransTokenizer

BATCH_SIZE = 4
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
quantization = None
```

```python
def initialize_model_and_tokenizer(ckpt_dir, direction, quantization):
    if quantization == "4-bit":
        qconfig = BitsAndBytesConfig(
            load_in_4bit=True,
            bnb_4bit_use_double_quant=True,
            bnb_4bit_compute_dtype=torch.bfloat16,
        )
    elif quantization == "8-bit":
        qconfig = BitsAndBytesConfig(
            load_in_8bit=True,
            bnb_8bit_use_double_quant=True,
            bnb_8bit_compute_dtype=torch.bfloat16,
        )
    else:
        qconfig = None

    tokenizer = IndicTransTokenizer(direction=direction)
    model = AutoModelForSeq2SeqLM.from_pretrained(
        ckpt_dir,
        trust_remote_code=True,
        low_cpu_mem_usage=True,
        quantization_config=qconfig,
    )

    if qconfig == None:
        model = model.to(DEVICE)
        if DEVICE == "cuda":
            model.half()

    model.eval()

    return tokenizer, model


def batch_translate(input_sentences, src_lang, tgt_lang, model, tokenizer, ip):
    translations = []
    for i in range(0, len(input_sentences), BATCH_SIZE):
        print(i)
        batch = input_sentences[i : i + BATCH_SIZE]

        # Preprocess the batch and extract entity mappings
        batch = ip.preprocess_batch(batch, src_lang=src_lang, tgt_lang=tgt_lang)

        # Tokenize the batch and generate input encodings
        inputs = tokenizer(
            batch,
            src=True,
            truncation=True,
            padding="longest",
            return_tensors="pt",
            return_attention_mask=True,
        ).to(DEVICE)

        # Generate translations using the model
        with torch.no_grad():
            generated_tokens = model.generate(
                **inputs,
                use_cache=True,
                min_length=0,
                max_length=256,
                num_beams=5,
                num_return_sequences=1,
            )

        # Decode the generated tokens into text
        generated_tokens = tokenizer.batch_decode(generated_tokens.detach().cpu().tolist(), src=False)

        # Postprocess the translations, including entity replacement
        translations += ip.postprocess_batch(generated_tokens, lang=tgt_lang)

        del inputs
        torch.cuda.empty_cache()

    return translations
```

```python
### English to Indic Example


en_indic_ckpt_dir = "ai4bharat/indictrans2-en-indic-1B"  # ai4bharat/indictrans2-en-indic-dist-200M
en_indic_tokenizer, en_indic_model = initialize_model_and_tokenizer(en_indic_ckpt_dir, "en-indic", quantization)

ip = IndicProcessor(inference=True)

# en_sents = [
#     "When I was young, I used to go to the park every day.",
#     "He has many old books, which he inherited from his ancestors.",
#     "I can't figure out how to solve my problem.",
#     "She is very hardworking and intelligent, which is why she got all the good marks.",
#     "We watched a new movie last week, which was very inspiring.",
#     "If you had met me at that time, we would have gone out to eat.",
#     "She went to the market with her sister to buy a new sari.",
#     "Raj told me that he is going to his grandmother's house next month.",
#     "All the kids were having fun at the party and were eating lots of sweets.",
#     "My friend has invited me to his birthday party, and I will give him a gift.",
# ]

src_lang, tgt_lang = "eng_Latn", "guj_Gujr"
print("startd")
eng_to_gu_mbart = batch_translate(english_sentences, src_lang, tgt_lang, en_indic_model, en_indic_tokenizer, ip)

# print(f"\n{src_lang} - {tgt_lang}")
# for input_sentence, translation in zip(en_sents, hi_translations):
#     print(f"{src_lang}: {input_sentence}")
#     print(f"{tgt_lang}: {translation}")

# flush the models to free the GPU memory
del en_indic_tokenizer, en_indic_model
```

## **Understanding BLEU and ROUGE Scores:**

### **BLEU (Bilingual Evaluation Understudy):**

- Measures the similarity between the generated translation and one or more reference translations based on n-gram overlap.
- Higher BLEU scores indicate better translation quality, focusing on the precision of n-gram matches.
- BLEU score emphasizes the literal word-for-word overlap between the generated text and the reference text.
- Limitations include not capturing semantic similarity and being sensitive to word order.


### **ROUGE (Recall-Oriented Understudy for Gisting Evaluation):**
- It is a set of metrics used for evaluating the quality of text summaries or translations by comparing them to reference texts. It focuses on measuring the overlap of n-gram sequences and word sequences between the generated text and the reference text.

####**ROUGE scores include:**

- **ROUGE-1**: Measures unigram overlap, capturing how many unigrams (individual words) in the generated text match those in the reference text.
- **ROUGE-2**: Measures bigram overlap, assessing how many bigram sequences (pairs of adjacent words) are shared between the generated text and the reference text.
- **ROUGE-L**: Computes the longest common subsequence (LCS) of words between the generated text and the reference text, emphasizing longer word sequences.

####**Each ROUGE score consists of precision (p), recall (r), and F1-score (f):**

- **Precision (p)**: Indicates the proportion of generated n-grams that appear in the reference text.
- **Recall (r)**: Reflects the proportion of reference n-grams that are present in the generated text.
- **F1-score (f)**: Harmonic mean of precision and recall, providing a balanced measure of how well the generated text matches the reference text.
